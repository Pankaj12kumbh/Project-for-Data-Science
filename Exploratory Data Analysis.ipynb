# Treatment of null Values

```python
#Treatment of null values by replacing it with 0
df['alcohol'].fillna(0,inplace=True)

#Treatment of null values by replacing it with mean values
df['alcohol'].fillna(df.alcohol.mean(),inplace=True)

#Treatment of null values by droping the row
df['alcohol'].dropna()
```

```python
df.rename(columns={"od280/od315_of_diluted_wines":"Diluted_value"}, inplace = True)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics

mean = statistics.mean(df['magnesium'])
sd = statistics.stdev(df['magnesium'])
  
plt.plot(df['magnesium'], norm.pdf(df['magnesium'], mean, sd))
plt.show()
```

```python
x_axis = np.arange(-20, 20, 0.01)
  
# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
print(sd)
  
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```
